```python
df = pd.read_csv('Label_pajak.csv')
```

```python
df['content'] = df['content'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df.head()
```

```python
df['content'] = df['content'].str.replace(r'[^\x00-\x7F]+', '', regex=True)

df.head(30)
```

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

factory = StopWordRemoverFactory()
stopwords_id = factory.get_stop_words()

def remove_stopwords(text):
  """custom function to remove the stopwords"""
  return " ".join([word for word in str(text).split() if word not in stopwords_id])

df['content'] = df['content'].apply(lambda text:remove_stopwords(text))
df.head()
```

```python
from Sastrawi.Stemmer.Stemmer import Stemmer

from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
Stemmer_id = factory.create_stemmer()

def stem_words(text):
  return " ".join([Stemmer_id.stem(word) for word in text.split()])

df['content'] = df['content'].apply(lambda text:stem_words(text))
df.head()
```

```python
def tokenize_text(text):
  tokenized_text = text.split()
  return tokenized_text
df['content'] = df['content'].apply(lambda x: tokenize_text(x))
df.head()
```

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
def lemmatize_text(token_list):
  return " ".join([lemmatizer.lemmatize(token) for token in token_list if not token in set()])

lemmatizer = nltk.stem.WordNetLemmatizer()
df['content'] = df['content'].apply(lambda x: lemmatize_text(x))
df.head()
```
